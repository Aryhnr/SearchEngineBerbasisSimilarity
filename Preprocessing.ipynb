```python
import json
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Load data JSON
with open('hasil/data_berita_final.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

data
```

```python
# Stopword Bahasa Indonesia
stop_words = set(stopwords.words('indonesian'))

# Inisialisasi stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
# Fungsi preprocessing per berita
def preprocess_text(text):
    # Lowercase
    text = text.lower()
    # Hilangkan URL
    text = re.sub(r"http\S+", "", text)
    # Hilangkan karakter non-alfabet
    text = re.sub(r"[^a-z\s]", "", text)
    # Hapus spasi berlebih
    text = re.sub(r"\s+", " ", text).strip()
    
    # Tokenisasi
    tokens = word_tokenize(text)
    
    # Stopword removal
    tokens = [word for word in tokens if word not in stop_words]
    
    # Stemming
    stemmed = stemmer.stem(" ".join(tokens))
    
    return stemmed
```

```python
# Proses semua data
for item in data:
    item['judul_bersih'] = preprocess_text(item['judul'])
    item['konten_bersih'] = preprocess_text(item['konten'])
```

```python
# Simpan ke JSON
with open('data_berita_bersih.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, indent=2, ensure_ascii=False)

print("Data sudah disimpan ke 'data_berita_bersih.json'")
```
